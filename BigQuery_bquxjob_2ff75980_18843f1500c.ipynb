# Result set loaded from BigQuery job as a DataFrame
Query results are referenced from the Job ID ran from BigQuery and the query
does not need to be re-run to explore results. The ```to_dataframe```
[method](https://googleapis.dev/python/bigquery/latest/generated/google.cloud.bigquery.job.QueryJob.html#google.cloud.bigquery.job.QueryJob.to_dataframe)
downloads the results to a Pandas DataFrame by using the BigQuery Storage API.

To edit query syntax, you can do so from the BigQuery SQL editor or in the
```Optional:``` sections below.

````python
# Running this code will read results from your previous job

job = client.get_job('bquxjob_2ff75980_18843f1500c') # Job ID inserted based on the query results selected to explore
results = job.to_dataframe()
results
````

````python
# Running this code will read results from your previous job

job1 = client.get_job('bquxjob_5199f4b4_18843f28be6') # Job ID inserted based on the query results selected to explore
results1 = job1.to_dataframe()
results1
````

````python
results=results.rename(columns={"InvoicedCompany_CompanyBkey": "CompanyBkey"})
````

````python
results.info()
````

````python
data_merged_df_book=results.merge(results1, on='CompanyBkey', how='left')
````

````python
data_merged_df_book.info()
````

````python
sub_data_merge=data_merged_df_book[["CompanyBkey", "RepeatCompany","Reporting_Group_Name__c"]]
````

````python
sub_data_merge.groupby(["Reporting_Group_Name__c", "RepeatCompany"], as_index=False)["CompanyBkey"].count()
````

````python
# Grouping by CompanyBkey
df_Preprocess = sub_data_merge.groupby(['RepeatCompany','Reporting_Group_Name__c']).agg({
        'CompanyBkey': 'count'})
````
